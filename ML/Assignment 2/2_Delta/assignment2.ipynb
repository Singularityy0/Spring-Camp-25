```python
import math
def inverse_square(x):
    return (1/(x*x))
plt.scatter(df['d'].apply(inverse_square), df['E'])
```

```python
df['Ed2'] = df['E']*df['d']*df['d']
```

```python
plt.scatter(df['B'], df['Ed2'])
```

```python
plt.scatter(df['Q'], df['Ed2'])
```

```python
def dTtransform2(x):
    return x**0.8
    

plt.scatter(df['dT'].apply(dTtransform2), df['Ed2'])
```

```python
df['QB'] = df['Q']*df['B']
```
